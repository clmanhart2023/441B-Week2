```python
# Make Region Binary 

## Making  Northwest Column
df.loc[df['region'] == 'northwest', 'northwest'] = 1.
df.loc[df['region'] == 'southwest', 'northwest'] = 0.
df.loc[df['region'] == 'northeast', 'northwest'] = 0.
df.loc[df['region'] == 'southeast', 'northwest'] = 0.

## Making  SouthWest Column
df.loc[df['region'] == 'northwest', 'southwest'] = 0.
df.loc[df['region'] == 'southwest', 'southwest'] = 1.
df.loc[df['region'] == 'northeast', 'southwest'] = 0.
df.loc[df['region'] == 'southeast', 'southwest'] = 0.


## Making  Northeast Column
df.loc[df['region'] == 'northwest', 'northeast'] = 0.
df.loc[df['region'] == 'southwest', 'northeast'] = 0.
df.loc[df['region'] == 'northeast', 'northeast'] = 1.
df.loc[df['region'] == 'southeast', 'northeast'] = 0.


## Making  Southeast Column
df.loc[df['region'] == 'northwest', 'southeast'] = 0.
df.loc[df['region'] == 'southwest', 'southeast'] = 0.
df.loc[df['region'] == 'northeast', 'southeast'] = 0.
df.loc[df['region'] == 'southeast', 'southeast'] = 1.

df
```

```python
# Rearrange the Data 
df = df[['age', 'sex', 'bmi', 'children', 'smoker',
       'northwest', 'southwest', 'northeast', 'southeast', 'charges']]
df
```

# 2.) Split the data into 80/20, in/out sample

```python
import numpy as np
```

```python
# Make Data & Target Array
data = np.array(df.iloc[:, :-1])
target = np.array(df.iloc[:, -1])
```

```python
# Determining Where to Cut
cut = int((len(data) * 0.8) // 1)
```

```python
# Split into In/Out Sample Data:
in_data = data[:cut]
out_data = data[cut:]

# Split into In/Out Sample Target:
in_target = target[:cut]
out_target = target[cut:]
```

# 3.) Normalize the Data

```python
from sklearn import preprocessing
```

```python
### In Sample Data
# Defining the Scaler
scaler = preprocessing.StandardScaler().fit(in_data)

# Converting in_data to be scaled
in_data_scale = scaler.transform(in_data)


### Out-of Sample Data
out_data_scale = scaler.transform(out_data)
```

# 4.) Get lambda from Lasso cross validation

```python
# Use function LassoCV
from sklearn.linear_model import LassoCV

modCV = LassoCV().fit(in_data_scale, in_target)
```

```python
# Determinging the optimal lambda
a = modCV.alpha_
a
```

# 5.) Run a lambda regression with that Lambda

```python
# Use the Lasso Function
from sklearn.linear_model import Lasso

mod1 = Lasso(alpha = a).fit(in_data_scale, in_target)
```

# 6.) Visualize the coefficients 

```python
# get the Coefficients
coefs = mod1.coef_
```

```python
l = list(zip(coefs, df))
coef_name = []
coef = []
for i in range(len(l)):
  coef_name.append(l[i][1])
  coef.append(l[i][0])


coefficients = pd.DataFrame(coef, index = coef_name, columns = ['Coefficient'])
coefficients
```

```python
# Plotting the Coefficients
import matplotlib.pyplot as plt
plt.figure(figsize = (15,8))

plt.bar(coefficients.index, coefficients['Coefficient'])
plt.axhline(0, color = 'red')
plt.title('Lasso Regression Output')
plt.xlabel('Beta')
plt.show()
```

# 7.) Interpret the coefficients

The chart above highlights that the smoker, age and bmi features are the most important for the regression. The coefficients are difficult to interpret directly because we need to "unnormalize" them

# 8.) Compare in and out of sample MSE’s

```python
### In Sample MSE
from sklearn.metrics import mean_squared_error
in_pred = mod1.predict(in_data_scale)

MSE = mean_squared_error(in_target, in_pred)

print('MSE:', MSE)
```

```python
### Out of Sample MSE
out_pred = mod1.predict(out_data_scale)

MSE = mean_squared_error(out_target, out_pred)

print('MSE:', MSE)
```
